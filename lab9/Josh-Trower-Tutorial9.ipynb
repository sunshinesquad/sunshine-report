```python
#importing used libraries
import csv
import re
from functools import reduce
import numpy as np
import pandas as pd

# this aux function reads the CSV file and returns the data in a Python dictionary
def get_data_csv(link):
    collection = []
    with open(link, 'r') as f:
        for line in csv.DictReader(f):
            collection.append(line)
        return collection

# the data for 2020's sunshine list
sunshine_2020 = pd.DataFrame(get_data_csv("sunshine2020.csv"))
sunshine_2020["Salary paid"] = sunshine_2020["Salary paid"].replace('[\$,]', '', regex=True).astype(float)
sunshine_2020["Taxable benefits"] = sunshine_2020["Taxable benefits"].replace('[\$,]', '', regex=True).astype(float)
#####Which sector has the most people/largest budget?#####
print(len(sunshine_2020))
sunshine_2020.head(5)
```

```python
sectors_money = sunshine_2020.groupby('Sector')['Salary paid'].sum().reset_index()
sectors_money.sort_values(by=["Salary paid"], ascending=False)
sectors_money.plot(title="Sum of all salarys",kind='bar', figsize=(12,6),x="Sector")
#THE MUNICIPALITIES SECTOR HAS THE LARGEST BUDGET
```

```python
sectors_taxable = sunshine_2020.groupby('Sector')['Taxable benefits'].sum().reset_index()
sectors_taxable.sort_values(by=["Taxable benefits"], ascending=False)
sectors_money.plot(title="Sum of all Taxable benefits",kind='bar', figsize=(12,6),x="Sector")
```
